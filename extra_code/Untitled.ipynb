```python
alcohol_file="patents_data_Alcohol.csv"
alcohol_dir="alcohol"
nalcohol_file="patents_data_Non-Alcohol.csv"
nalcohol_dir="nalcohol"
auto_file="patents_data_Autonomous.csv"
auto_dir="auto"
nauto_dir="nauto"
nauto_file="patents_data_Non_Autonomous.csv"
head_url='https://patents.google.com/patent/'
```

```python
import os
with open(alcohol_file,'r') as csvfile:
    os.mkdir(alcohol_dir)
    reader=csv.reader(csvfile)
    for row in reader:
        patent_num=str(row[0])
        url=head_url+patent_num
        try:
            page = requests.get('https://patents.google.com/patent/'+patent_num) 
        except:
            print("A connection error!"+patent_num)
            continue
        data = page.text
        soup= bsp(data,'html.parser')
        results = soup.findAll("meta", {"name" : "description"})
        description=results[0]["content"]
        description=description.replace('\n',' ')
        description=description.replace('  ',' ')
        results = soup.findAll("span", {"itemprop" : "Description"})
        title=results[-1].text
        to_write=title+' '+description
        write_file= open(alcohol_dir+'/'+patent_num+".txt","w+")
        write_file.write(to_write)
        write_file.close()
        
        
        
    
```
